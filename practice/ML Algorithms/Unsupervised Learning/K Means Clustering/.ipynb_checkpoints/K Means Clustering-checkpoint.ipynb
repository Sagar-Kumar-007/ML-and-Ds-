```python
# The process to form clusters:
```

```python
# First take random "K" points on the graph and say them as the centre of the cluster (Centroid).
Image("centroids.png",width=700)
```

```python
# The centroids could be placed anywhere in the graph.
```

```python
# Second step is to find the distances of the data points to the centroids. The data points that are near to Centroid (lets
# say "A") form cluster "A" and the data points that are near to Centroid (lets say "B") form cluster "B". Easy method to 
# divide them according to the nearest distance is to draw a perpendicular between the centroids. The data points that
# lie on the side of Centroid "A" are part of cluster "A" and vice versa.
Image("formation.png",width=700)
```

```python
# There you go. We got two improper clusters. Now, our task is to improve these clusters.
```

```python
# Third step is to put these centroids to their original places i.e., for the red cluster you take the centroid of the four
# red data points and put the red Centroid there and repeat the process for green centroid.
Image("position.png",width=700)
```

```python
# Next step is to again find the nearest distances between the data points and the centroid. Now if the green data points
# are near to the red centroid then change them to red and repeat the same process for green centroid.
Image("formation_2.png",width=700)
```

```python
# Now again you find the centroids of the clusters and place their respective centroids and find the nearest distances between
# them and reconstruct the clusters. Repeat the whole procedure until the position of the centroid doesnot change the cluster
# construction. Finally you will get the required clusters.
```

# How to determine Correct Number of CLusters (K):

```python
# The method to determine the correct number of clusters is called Elbow Method.
```

```python
# In elbow method, first you measure the sum of squares of distances between the centroid and the data points of the respective
# clusters (in short you are finding the error between the data points and their centroids) (SSE) and then you add all the 
# SSE's of the clusters. Since, the number of clusters are "K", you will get "K" number of SSE's.
Image("SSE.png",width=700)
```

```python
# Here xi is the position of the data point and ci is the centroid.
# SSE= sum of individual SSE's
```

```python
# You find the SSE for various values of "K" and then plot them against their respective SSE.
Image("elbow.png",width=700)
```

```python
# After plotting the graph, you will see that at a point the SSE will become zero. This is so because now the clusters are
# containing individual data points with their respective centroids (at same position) and hence the error becomes zero.
# You can select the best value of "K" as the point where the graph gets a slight bump similar to that of an elbow.
# Since, in the above graph, the bump is at K=4, we are assuming that our data can be clustered into 4 different clusters.
```

```python
import pandas as pd
```

```python
df=pd.read_csv("income.csv")
df
```

```python
# From the above dataset we can say that in a company there are few people with their respective age and salaries.
```

```python
import matplotlib.pyplot as plt
```

```python
# Lets plot a scatter plot among its features to have a visual representation of the data.
plt.scatter(df["Age"],df["Income($)"])
```

```python
# You can see that there can a formation of three clusters. Hence, for this problem, choosing "K" would be pretty\
# straightforward.
```

```python
# Lets import KMeans from sklearn.cluster.
from sklearn.cluster import KMeans
```

```python
# Now this ml model does is that while training itself on the features it divides the dataset into clusters and then while
# predicting, it predicts whether the test data will belong to which cluster.
```

```python
km=KMeans(n_clusters=3)
```

```python
km.fit(df[["Age","Income($)"]])
```

```python
cluster_no=km.predict(df[["Age","Income($)"]])
cluster_no
```

```python
# Now you can say by looking at the above array that which person belonged to which cluster.
```

```python
cluster=pd.DataFrame()
cluster["Name"]=df["Name"]
cluster["cluster_number"]=cluster_no
cluster
```

```python
# Now lets visualize these clusters in a scatterplot
```

```python
# First lets append the predicted clusters to the original dataframe and divide them according to their clusters.
```

```python
df["cluster"]=cluster_no
df
```

```python
df1=df[df["cluster"]==0]
df2=df[df["cluster"]==1]
df3=df[df["cluster"]==2]
```

```python
# Now we plot a scatter plot among them:
plt.scatter(df1["Age"],df1["Income($)"])
plt.scatter(df2["Age"],df2["Income($)"])
plt.scatter(df3["Age"],df3["Income($)"])
plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

```python
# Now you can see the clusters that were divided.
```

```python
# You can see that the yellow cluster is seperated properly but the green and the blue are not clustered properly. The reason
# is so because the income scale is 20,000 and that of Age is 2.5 ... Hence we need to feature scale the data before clustering.
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
# We are using MinMaxScaler for this purpose because age and income cannot be negative.
```

```python
scaler.fit(df[["Income($)"]])
```

```python
df["Income($)"]=scaler.transform(df[["Income($)"]])
```

```python
df
```

```python
# Lets scale the age
```

```python
scaler.fit(df[["Age"]])
df["Age"]=scaler.transform(df[["Age"]])
df
```

```python
# Now we again cluster them using KMeans and then make a scatter plot
```

```python
km.fit(df[['Age','Income($)']])
```

```python
km.predict(df[['Age','Income($)']])
```

```python
df["cluster"]=km.predict(df[['Age','Income($)']])
```

```python
df
```

```python
# Now we divide them into three different datasets
df1=df[df["cluster"]==0]
df2=df[df["cluster"]==1]
df3=df[df["cluster"]==2]
```

```python
# Now we plot a scatter plot among them:
plt.scatter(df1["Age"],df1["Income($)"])
plt.scatter(df2["Age"],df2["Income($)"])
plt.scatter(df3["Age"],df3["Income($)"])
plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

```python
# You can also get the position of the centroids for the respective clusters:
km.cluster_centers_
```

```python
# Lets visualize the centroids int the above scatterplot:
plt.scatter(df1["Age"],df1["Income($)"])
plt.scatter(df2["Age"],df2["Income($)"])
plt.scatter(df3["Age"],df3["Income($)"])
plt.scatter(x=km.cluster_centers_[:,0],y=km.cluster_centers_[:,1],marker="+",s=100)
plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

```python
# The "+" signs are the respective centroids for the clusters.
```

```python
# Lets use Elbow method to find the best "K":
```

```python
# We use KMeans().inertia_ to find the SSE (sum of all the individual SSE's).
```

```python
SSE=[]
k_range=range(1,20) # We are taking range of "K" values.
for k in k_range:
    km=KMeans(n_clusters=k) # We want "K" clusters
    # Now we fit the model and then calculate the SSE and append it to a list
    km.fit(df[['Age','Income($)']])
    SSE.append(km.inertia_)
```

```python
SSE
```

```python
# Now we plot a graph between "K" and SSE:
```

```python
plt.plot(k_range,SSE)
plt.xlabel("K")
plt.ylabel("SSE")
plt.show()
```

```python
# Since, we are not able to find the x-value corresponding to the elbow, lets shorten the x-axis:
plt.plot(k_range[0:10],SSE[0:10])
plt.xlabel("K")
plt.ylabel("SSE")
# plt.show()
```

```python
# We repeat the process again
plt.plot(k_range[0:5],SSE[0:5])
plt.xlabel("K")
plt.ylabel("SSE")
plt.show()
```

```python
# You can see that the value of "K" from the elbow method is 3.
```

```python
# Elbow method is useful in finding "K" when there are multiple features and the scatterplot looks clumsy.
```
