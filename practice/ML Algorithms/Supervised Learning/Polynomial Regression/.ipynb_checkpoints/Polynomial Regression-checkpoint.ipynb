```python
from IPython.display import Image
```

```python
# When a plot against independent variables and dependent variable looks more like that of a curve than that of a line,
# we use Polynomial Regression.
```

```python
Image(filename="Polynomial Regression.png",width=500)
```

```python
# You can see that the dots tends to form a curve than a line. Hence, in such cases we use Polynomial Regression rather than
# Linear Regression.
```

```python
# Make sure to adjust the degree of the polynomial such that the model doesnot overfit or underfit the model.
# Overfitting can be achieved by increasing the degree of the polynomial than required.
# Underfitting can be achieved by decreasing the degree of the polynomial than required.
```

```python
# Polynomial Regression takes time to train when you have large datasets, and hence it is generally not used.
# But using Polynomial Regression increases the accuracy of the model.
```

```python
# If the degree of the polynomial is greater than required, your plot would apper to look like the one below (overfitting).
Image(filename="Overfitting.png",width=500)
```

```python
# If the degree of the polynomial is less than required, your plot would look like the one below (underfitting).
Image(filename="Underfitting.png",width=500)
```

```python
# The first image in this notebook is the Generalized fit. We train the model such that it would achieve Generalized fitting.
```

```python
# The error/residual would be:
# Underfit: Max
# Overfit: Zero
# Generalized Fit: In between 0 and max.
```

```python
# We train the model to achieve Generalized fit and not overfit because:
# Lets take a new independent variable on the graph for both overfit and generalized fit, you may see that the error between
# the actual value and the predicted value would be less for the generalized fit rather than the overfit.
Image(filename="generalized fit test.png",width=500)
```

```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
df = pd.read_csv(path)
```

```python
pd.set_option("display.max_columns",None)
```

```python
df.head()
```

```python
X=df.drop(columns="price")
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
std_scaler.fit(X_train)
```

```python
X_train=std_scaler.transform(X_train)
X_test=std_scaler.transform(X_test)
```

# Polynomial Regression: Training and testing the model

```python
# To perform polynomial regression, you first convert the X_train and X_test data and give them the polynomial features
# using PolynomialFeatures class of scikit-learn and then perform Linear Regression to them.
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# First we convert our data so that it contains polynomial features.
```

```python
# We create an object having the PolynomialFeatures class
poly_features=PolynomialFeatures(degree=2)
```

```python
# We fit our model
poly_features.fit(X_train)
```

```python
# Then we Transform the features to the dataset
X_train_poly=poly_features.transform(X_train)
X_test_poly=poly_features.transform(X_test)
```

```python
X_train.shape,X_test.shape,X_train_poly.shape,X_test_poly.shape
```

```python
# You can see that the data got the polynomial features.
```

```python
# Now we perform Linear Regression to the data
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train_poly,y_train)
```

```python
lr.score(X_test_poly,y_test)
```

```python
# You can observe that we get a score of about 99.93% which is very high. The score is high but that does not mean 
# your model can be deployed. For this, you must check whether your model is generalized fit or overfit or underfit. If
# the model is generalized fit then you can deploy your model. If the model is overfit then you must
# decrease the degree of the polynomial
```

```python
y_pred=lr.predict(X_test_poly)
```

```python
y_pred
```

```python
y_test
```

```python
# You can object that both the dataset and the array contains nearly/exactly the same values. This may happen if your
# model is trained well or else if your model is overfit.
```

```python
# Now we calculate the root mean squared error 
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
rmse=mse**0.5
rmse
```

```python
# You can see that the rmse is very low and hence there is very liitle error in our model. Hence, it can be one of
# the best model.
```
