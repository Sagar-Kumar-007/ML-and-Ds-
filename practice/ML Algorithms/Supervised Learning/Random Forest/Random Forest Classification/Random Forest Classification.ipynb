# Working Principle:

```python
Image(filename="datasets.png", width=700)
```

```python
# In the above image, the dataset on the left is the original dataset.From the original dataset, we select random rows and
# columns (making sure that the size of the original dataset is greater than the size of the derived dataset) and split the
# original dataset into multiple derived datasets. It depends on us the number of decision trees to be created.
```

```python
Image(filename="working.png", width=600)
```

```python
# Now we train each derived dataset and make a corresponding decision tree. Then we predict the values.
# While predicting the values, the majority class (i.e., the output which has occured most frequent) becomes the final-class
# and is the predicted value for that test data.
```

```python
# In classification, we consider the most frequent value but in regression, since the output of each decision tree is
# a continuous value, we take the average of all the outputs and that becomes the predicted value for the test data.
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
data
```

```python
data.keys()
```

```python
data.feature_names
```

```python
data.data
```

```python
data.target
```

```python
data.target_names
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
df["target"]=data.target
```

```python
df.head()
```

```python
# Now we split our data into features and target
X=df.drop(columns="target")
y=df["target"]
```

```python
# Now we split data into training and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Now we train the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier(n_estimators=100,criterion='gini')
# n_estimators is used to create "n" number of decision trees.
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
# Now we predict whether a patient is suffering from cancer
```
