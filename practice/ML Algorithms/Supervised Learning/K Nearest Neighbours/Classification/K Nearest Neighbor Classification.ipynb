```python
# Suppose that we have trained our model using KNN for this dataset and we are testing our data,
# how will the model predict whether the test data belong to Malignant or Benign Tumor?
```

```python
# To answer the above question, we can follow the steps given below:
Image(filename="steps.png",width=700)
```

```python
# First step is that we have to get labeled data (data which contain dependent variables).
```

```python
# Second step is to define the "K" in KNN.
# "K" is "5" by default. Make sure that the value of "K" is odd. "K" must be odd in order to get the majority of the following
# neighbours.
# A good rule of thumb is that K=sqrt(N), where N = no. of data samples.
```

```python
# Third step is to calculate the distance from the test data point to all the data points and select the nearest
# "K" data points (neighbours).
```

```python
# Fourth step is to count the number of neighbours belonging to each class and then select the neighbour with highest count
# i.e., the class with majority.
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
data
```

```python
df=pd.DataFrame(data=data.data,columns=np.array(data.feature_names))
df.head()
```

```python
df["target"]=data.target
```

```python
# Now we split the dataset into independent and dependent variables.
X=df.drop(columns="target")
y=df["target"]
```

```python
# Now we split them into training and test data
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
# Now we train the model
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=21)
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
# Now we predict whether a patient is suffering from cancer.
```
