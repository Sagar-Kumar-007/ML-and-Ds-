```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
df = pd.read_csv(path)
pd.set_option("display.max_columns",None)
```

```python
X=df.drop(columns="price")
```

```python
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
# Now we train the model:
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
regressor=KNeighborsRegressor(n_neighbors=5)
```

```python
regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

```python
regressor.predict(X_test)
```

```python
# Lets test the score for feature scaled data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
std_scaler.fit(X_train)
X_train_std=std_scaler.transform(X_train)
X_test_std=std_scaler.transform(X_test)
```

```python
regressor_std=KNeighborsRegressor()
```

```python
regressor_std.fit(X_train_std,y_train)
```

```python
regressor_std.score(X_test_std,y_test)
```

```python
regressor_std.predict(X_test_std)
```

```python
y_test
```

```python
# It appears that our model appears to be more accurate without feature scaling.
```
