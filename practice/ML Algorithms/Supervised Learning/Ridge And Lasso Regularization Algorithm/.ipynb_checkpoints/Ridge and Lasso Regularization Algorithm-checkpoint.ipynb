```python
# Linear regression draws a line which has minimum squared difference between predicted value and original value (i.e., ei).
from IPython.display import Image
from IPython.core.display import HTML
Image(filename="linear regression.png",width=500)
```

```python
# Due to this there are chances when the line overfits the data and hence resulting in test errors.
```

# Lasso Regression

```python
Image(filename="lasso regression.png",width=800)
```

```python
# You can observe that in Lasso Regression we have added a term lambda to the ei (i.e., the square of the predicted value 
# subtracted from actual value). The lambda is called as alpha in scikit-learn. During Linear regression we try to 
# make the predicted value as close as the actual value hence turning the ei value to be zero and hence overfitting the 
# data. To resolve this we generalize the expression by adding a term in which alpha (generally one, but can be changed)
# is multiplied by modulus of the slope of the line. This prevents the overfitting of the model.
```

# Ridge Regression

```python
Image(filename="ridge regression.png",width=800)
```

```python
# The reason to use ridge regression is also the same as that of lasso regression. The only difference is that in 
# ridge regression the slope is squared and in lasso regression the slope is in modulus.
```

```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
df = pd.read_csv(path)
pd.set_option("display.max_columns",None)
```

```python
df.head()
```

```python
X=df.drop(columns="price")
```

```python
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
std_scaler.fit(x_train)
```

```python
x_train=std_scaler.transform(x_train)
x_test=std_scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Now tha we have our x_train and x_test data standardized we can train our model.
```

```python
# First we train our model with Linear Regression and then with Ridge and Lasso Regression to compare the scores.
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
# Now we implement Ridge Regression
r_reg=Ridge()
r_reg.fit(x_train,y_train)
r_reg.score(x_test,y_test)
```

```python
# We can observe that the model's score is 79.03% while using Linear Regression and 79.05% while using Ridge Regression 
# although the model's score is not increased significantly.
```

```python
# Now we try Lasso Regression
l_reg=Lasso()
l_reg.fit(x_train,y_train)
l_reg.score(x_test,y_test)
```

```python
# You can see that the score of the model is increased by 1% which is a significant increase.
```

```python
# Now lets change the values of alpha to see the impact on the score
```

```python
r_reg2=Ridge(alpha=2)
r_reg2.fit(x_train,y_train)
r_reg2.score(x_test,y_test)
```

```python
# You can see there is a little increase in the score. Now lets check the score using lasso
l_reg2=Lasso(alpha=2)
l_reg2.fit(x_train,y_train)
l_reg2.score(x_test,y_test)
```

```python
# You can see there is a significant increase in the score.
```

```python
l_reg3=Lasso(alpha=3)
l_reg3.fit(x_train,y_train)
l_reg3.score(x_test,y_test)
```

```python
# You can see there is a significant increase in the score.
```

```python
l_reg100=Lasso(alpha=100)
l_reg100.fit(x_train,y_train)
l_reg100.score(x_test,y_test)
```

```python
# You can see that there is a significant decrease in the score. Hence we can conclude that the score has a large impact
# when you can change the value of alpha. We find alpha (to get the best score) using a method called Hyper Parameter Tuning.
```
