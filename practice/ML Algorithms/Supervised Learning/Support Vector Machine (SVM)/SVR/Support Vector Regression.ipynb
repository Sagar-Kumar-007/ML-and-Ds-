```python
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
df.head()
```

```python
pd.set_option("display.max_columns",None)
```

```python
# First We split the independent and dependent columns.
X=df.drop(columns="price")
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Then we split the data.
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Then we perform feature scaling to the data.
std_scaler=StandardScaler()
```

```python
std_scaler.fit(X_train)
```

```python
X_train=std_scaler.transform(X_train)
X_test=std_scaler.transform(X_test)
```

```python
# Now we train our model.
```

```python
from sklearn.svm import SVR
```

```python
svr_rbf=SVR(kernel='rbf')
```

```python
svr_rbf.fit(X_train,y_train)
```

```python
svr_rbf.score(X_test,y_test)
```

```python
# You can see that our score is very low. Hence this model cannot be used to predict prices.
```

```python
svr_linear=SVR(kernel='linear')
svr_linear.fit(X_train,y_train)
svr_linear.score(X_test,y_test)
```

```python
# You can see that there is increase in our score.
```

```python
svr_poly=SVR(kernel='poly',degree=2)
svr_poly.fit(X_train,y_train)
svr_poly.score(X_test,y_test)
```

```python
# You can see that our score is reduced. To select the best parameters we use HyperParameterTuning concept.
```

```python
# Now we predict the values
y_pred=svr_linear.predict(X_test)
y_pred
```

```python
svr_linear.predict([X_test[0]])
```

```python
# We can also find the error using root mean square error
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=mse**0.5
rmse
```
