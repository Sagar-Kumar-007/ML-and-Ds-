# Mean Squared Error and Root Mean Squared Error

```python
Image(filename="rms vs ms error.png",width=500)
```

```python
# We generally use rms error while predicting the best model. Note that if the rms value is tending to zero (like 0,0.12
# etc.) we call the model to be overfit that is, our model has memorized all the values and cannot predict new values
# accurately. If the rms values are 100,200 etc. we call the model to be underfit. Hence we check the rms values of
# all our ML models and check for the least rms error making sure that it is not overfit.
```

```python
import pandas as pd
```

```python
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
df = pd.read_csv(path)
```

```python
X=df.drop(columns="price")
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
std_scaler.fit(X_train)
```

```python
X_train=std_scaler.transform(X_train)
X_test=std_scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

# Model Evaluation: RMS Error

## Method 1:

```python
y_pred=lr.predict(X_test) # Predicted values
y_pred
```

```python
# Actaul Values
y_test
```

```python
y_test-y_pred
```

```python
(y_test-y_pred)**2
```

```python
((y_test-y_pred)**2).sum()
```

```python
((y_test-y_pred)**2).sum()/y_test.shape[0]
```

```python
RMS_Error=(((y_test-y_pred)**2).sum()/y_test.shape[0])**0.5
RMS_Error
```

## Method 2:

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
print(mse)
```

```python
rmse=np.sqrt(mse)
print(rmse)
```

```python
# From the observations, we can say that the rmse is too high and hence we cannot use our model to predict the cost
# of house. Ex: Consider the prices to be in lakhs which results in the rmse to also be in lakhs. Suppose a house price
# is 100 lakhs then our model will predict the cost to be (100-65) 35 lakhs which is incorrect. Hence this model cannot
# be used to predict the price of a house
```

```python
# To rectify this we can perform data preprocessing in another way or we can change our model. But the above model cannot
# be deployed to predict the prices.
```
