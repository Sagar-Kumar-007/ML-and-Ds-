# Goodness of Fit:

```python
# It shows the closeness between the actual value and the predicted value.
```

```python
# We get the r-squared value using the score() method
```

```python
import pandas as pd
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
df = pd.read_csv(path)
```

```python
pd.set_option("display.max_columns",None)
```

```python
df.head()
```

```python
X=df.drop(columns="price")
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
```

```python
std_scaler.fit(X_train)
```

```python
X_train=std_scaler.transform(X_train)
X_test=std_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
y_pred=lr.predict(X_test)
```

# Implementing R-square from sklearn

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
# You can see that both the values from the score method and from the sklearn method are same
```
