```python
# You can fill missing values of categorical data using either:-
    # Global Constant:- You can fill the entire column's missing values as unknown,missing or similar to it.
    # You can fill them manually as per your domain knowledge.
    # Fill it with the value which repeats most frequently (mode).
```

```python
# You can use mode only when missing values in the dataset are random and the percentage of missing values is about 5%
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\\Users\\91830\\Downloads\\data science\\practice\\Data Cleaning\\house price dataset\\train.csv")
df
```

```python
df.info()
```

```python
df_categorical=df.select_dtypes(include=["object"])
```

```python
pd.set_option("display.max_columns",None)
```

```python
pd.set_option("display.max_rows",None)
```

```python
df_categorical.head()
```

```python
missing_percent=df_categorical.isnull().sum()/df_categorical.shape[0]*100
missing_percent
```

```python
plt.figure(figsize=[10,10])
sns.heatmap(df_categorical.isnull())
```

```python
missing_list=missing_percent[missing_percent>20].keys()
missing_list
```

```python
df_dropped=df_categorical.drop(missing_list,axis=1)
```

```python
plt.figure(figsize=[10,10])
sns.heatmap(df_dropped.isnull())
```

```python
df_dropped.shape
```

```python
# Now we should access the columns with missing values in order to fill them
```

```python
missing_percent=df_dropped.isnull().sum()/df_dropped.shape[0]*100
missing_percent
```

```python
missing_list=missing_percent[missing_percent>0].keys()
missing_list
```

```python
# Now you can fill them with global constant using loop...df_dropped['MasVnrType'].fillna("Missing")
# Or you can fill them with the mode of the column
```

```python
# You can see the count of each class in the column using value_counts() method
df_dropped['MasVnrType'].value_counts()
```

```python
# To get the mode of any column, you can
df_dropped['MasVnrType'].mode()
```

```python
type(df_dropped['MasVnrType'].mode())
# You can see that while using mode, you get a series...but you need to access the string of the series which is to be filled
# in place of the missing values...hence you specify the column using the index [0]
```

```python
df_dropped['MasVnrType'].fillna(df_dropped['MasVnrType'].mode()[0])
```

```python
# Now you have to fill the whole dataset with the mode of their columns. So we use a loop to do this
df_final=df_dropped.copy()
for var in missing_list:
    df_final.update(df_final[var].fillna(df_final[var].mode()[0]))
```

```python
df_final
```

```python
plt.figure(figsize=[10,10])
sns.heatmap(df_final.select_dtypes(include=["object"]).isnull())
```

```python
df_final.select_dtypes(include=["object"]).isnull().sum().sum()
```

```python
# Hence all the columns are filled
```

```python
# Lets compare the df_final and df histograms to know whether our cleaned dataset resembles the original dataset
```

```python
plt.figure(figsize=[16,9])
for i,var in enumerate(missing_list):
    plt.subplot(4,3,i+1)
    plt.hist(df_final[var],label="Cleaned",color="b")
    plt.hist(df[var].dropna(),label="Original",color="r")
    plt.legend()
```

```python
# You can observe that there are minute changes in the dataset...so our cleaned data may be accurate
```
