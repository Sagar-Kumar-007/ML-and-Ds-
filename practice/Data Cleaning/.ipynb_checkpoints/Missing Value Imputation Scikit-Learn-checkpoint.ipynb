# Missing values Imputation (Numeric datatype)

```python
num_vars=x_train.select_dtypes(include=["int64","float64"]).keys()
```

```python
num_vars
```

```python
# SimpleImputer method replaces the specified value (NaN by default) with a value of specified strategy (mean by default)
# We specify the SimpleImputer() to a variable which is then ready to be used.
imputer_mean= SimpleImputer(strategy="mean")
```

```python
# fit method is used to train the model...since SimpleImputer is used it will replace the missing values.
imputer_mean.fit(x_train[num_vars])
```

```python
# statistics_ method gives the value which is going to be replaced by the missing values. Here, it is the mean of the columns.
imputer_mean.statistics_
```

```python
x_train[num_vars].isnull().sum()
```

```python
# You will observe that the missing values are not yet replaced. This is because transform method isn't called yet.
```

```python
imputer_mean.transform(x_train[num_vars])
```

```python
# You observe that the output is an 2D array with all the missing values filled with their respective mean.
# Since the output is an 2D array, it is a disadvantage of using scikit-learn, but can convert the 2D array to a Dataset.
```

```python
x_train[num_vars]=imputer_mean.transform(x_train[num_vars])
```

```python
# Now we must also fill the test dataset's missing values with the mean of the columns of the train dataset and not test dataset
# Note that we use the values of train dataset to fill the test dataset. This is because we train the model using train dataset
# and not test dataset.
```

```python
# Let's first check its missing values. Even if there are no missing values, we do this process because may be  in future
# our test dataset may contain some missing values.Since our model is already trained (fit) to replace the missing values
# with the mean of x_train columns, we do not need to train the model again.
test[num_vars]=imputer_mean.transform(test[num_vars])
```

```python
x_train[num_vars].isnull().sum().sum()
```

```python
test[num_vars].isnull().sum().sum()
```

```python
# Hence all the missing values are replaced with the mean of x_train columns.
```

```python
# Now if you want to fill the missing values with a constant (say 1) value you can use:-
# imputer_mean= SimpleImputer(strategy='constant', value=1) and then fit the x_train data and use transorm to
# fill the missing values.
```

# Categorical Missing Value Imputation

```python
# Now we will use the same procedure but use the strategy as most_frequent (mode) instead of mean
```

```python
# Select a model
inpute_mode=SimpleImputer(strategy="most_frequent")
```

```python
cat_vars=x_train.select_dtypes(include=["object"]).keys()
cat_vars
```

```python
# Train the model
inpute_mode.fit(x_train[cat_vars])
```

```python
# Transform the model
x_train[cat_vars]=inpute_mode.transform(x_train[cat_vars])
test[cat_vars]=inpute_mode.transform(test[cat_vars])
```

```python
x_train[cat_vars].isnull().sum().sum()
```

```python
test[cat_vars].isnull().sum().sum()
```

```python
# Hence, missing values of categorical datatype are now filled.Lets check whether the complete training dataset and
# test dataset are now filled.
x_train.isnull().sum().sum()
```

```python
test.isnull().sum().sum()
```

```python
# Hence our complete training and test datasets are now filled.
```
